```python
# let's drop the redundant columns before EDA
df = df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1) 
```

```python
corr_matrix=df.select_dtypes(np.number).corr()  
fig, ax = plt.subplots(figsize=(15, 15))
ax = sns.heatmap(corr_matrix, annot=True)
plt.savefig("../images/output_corr.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
# JobLevel and MonthlyIncome are very highly correlated. So I drop JobLevel column
df = df.drop(['JobLevel'], axis=1) 
```
